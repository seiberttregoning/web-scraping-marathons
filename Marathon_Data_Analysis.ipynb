```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_2019 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 0)
full_2018 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 1)
full_2017 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 2)
full_2016 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 3)
half_2019 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 4)
half_2018 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 5)
half_2017 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 6)
half_2016 = pd.read_excel('data\marathon_data.xlsx', sheet_name = 7)
```

```python
full_2019['Time'] = full_2019['Time'].map(lambda x: pd.to_timedelta(x).seconds)
full_2018['Time'] = full_2018['Time'].map(lambda x: pd.to_timedelta(x).seconds)
full_2017['Time'] = full_2017['Time'].map(lambda x: pd.to_timedelta(x).seconds)
full_2016['Time'] = full_2016['Time'].map(lambda x: pd.to_timedelta(x).seconds)
half_2019['Time'] = half_2019['Time'].map(lambda x: pd.to_timedelta(x).seconds)
half_2018['Time'] = half_2018['Time'].map(lambda x: pd.to_timedelta(x).seconds)
half_2017['Time'] = half_2017['Time'].map(lambda x: pd.to_timedelta(x).seconds)
half_2016['Time'] = half_2016['Time'].map(lambda x: pd.to_timedelta(x).seconds)
```

```python
def full_histogram(dataframe):
    """Makes histograms out of a list of marathon dataframes"""
    
    plt.hist(dataframe['Time'], bins = 20, range = (9000, 27000))
    plt.xlabel('seconds')
    plt.ylabel('count')
    plt.title('Histogram of Time Results (without outliers)')
    
    return plt.show()
```

```python
full_marathons = [full_2019, full_2018, full_2017, full_2016]
```

```python
for i in full_marathons:
    full_histogram(i)
```

```python
def half_histogram(dataframe):
    """Makes histograms out of a list of marathon dataframes"""
    
    plt.hist(dataframe['Time'], bins = 20, range = (5000, 20000))
    plt.xlabel('seconds')
    plt.ylabel('count')
    plt.title('Histogram of Time Results (without outliers)')
    
    return plt.show()
```

```python
half_marathons = [half_2019, half_2018, half_2017, half_2016]
```

```python
for i in half_marathons:
    half_histogram(i)
```
